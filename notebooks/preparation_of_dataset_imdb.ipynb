# Aim is to create a csv file containing image id and gender from imdb.mat file 

```python
import pandas as pd
import scipy.io
```

```python
IMDB_DATASET_PATH = '../data/imdb_crop/'
```

```python
imdb_mat_file = IMDB_DATASET_PATH + 'imdb.mat'
mat = scipy.io.loadmat(imdb_mat_file)
```

```python
mat
```

```python
mat['imdb'][0][0]
```

```python
txt_path = IMDB_DATASET_PATH + 'mat.txt'
file1 = open(txt_path,"a") 
file1.write(str(mat))
file1.close()
```

```python
imdb_mat = mat['imdb']
```

```python
imdb_mat.dtype
```

```python
image_paths = imdb_mat['full_path'][0,0][0]
image_paths
```

```python
image_list = []
for image in image_paths:
    image_list.append(image[0])
```

```python
genders = imdb_mat['gender'][0,0][0]
genders
```

```python
gender_list = []
for gender in genders:
    gender_list.append(gender)
```

```python
op = []
for (image_path, gender) in zip(image_paths, genders): 
     op.append([image_path[0], gender])
```

```python
op
```

```python
gender_df = pd.DataFrame(op)
gender_df.columns = ['image_path', 'gender']
```

```python
csv_path = IMDB_DATASET_PATH + 'new_gender_df.csv'
gender_df.to_csv(csv_path,index=False)
```

```python
gender_df = pd.DataFrame({'image_path': image_list, 'gender': gender_list})
```
